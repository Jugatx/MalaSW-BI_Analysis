```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

# B2C Sales Visualisations

```python
# Plotting revenue for 2018, 2019 and 2020 from each of the different sale channels:
#Retail, wholesale, consignment, online and Caravana
#Note that the data for 2018 and 2019 comes from the financial reporting
#The data for 2020 comes from a sales file
#The latter file will need to be transformed to be able to plot 2020 along with 2018 and 2019
```

```python
#Importing file 1
df_total_ = pd.read_csv('../datasets/Sales/B2C/df_total_.csv', index_col=0)
df_total_
```

```python
#Importing file 2
df_2020 = pd.read_csv('../datasets/Sales/B2C/SalesB2C_2020.csv', index_col=0)
df_2020
```

```python
#Preparing DATA from 2020 for plot
```

```python
df_2020.columns
```

```python
df_2020.drop(['CLIENTE', 'CANTIDAD','MODELO','COLOR','TALLA','DESCUENTO','PRECIO'], axis=1, inplace=True)
```

```python
df_2020.info()
```

```python
df_2020['FECHA'].unique()
```

```python
df_2020= df_2020[~df_2020['FECHA'].isin(['?'])]
```

```python
df_2020['FECHA'] = pd.to_datetime(df_2020['FECHA'])
```

```python
#Extracting month from datetime
df_2020['MONTH'] = pd.DatetimeIndex(df_2020['FECHA']).month 
```

```python
df_2020
```

```python
#Mapping numerical data to categorical months for plot
month_map = {1:'Enero',2:'Febrero',3:'Marzo',4:'Abril',5:'Mayo',6:'Junio',
             7:'Julio',8:'Agosto',9:'Septiembre', 10:'Octubre',11:'Noviembre', 12 :'Diciembre'}
```

```python
df_2020['MONTH'].replace(month_map, inplace=True)
df_2020
```

```python
df_2020['MONTH'].unique()
```

```python
df_2020= df_2020[~df_2020['MONTH'].isin(['Diciembre'])]
```

```python
df_2020['PUNTO DE VENTA'].value_counts()
```

```python
df_2020= df_2020[~df_2020['PUNTO DE VENTA'].isin(['UNKNOWN'])]
```

```python
#Sales from instagram account for sales from the showroom as they were ordered via the instagram chat
df_2020['PUNTO DE VENTA'].replace('INSTAGRAM', 'SHOWROOM', inplace=True)
df_2020
```

```python
df_2020.drop(['FECHA'], axis=1, inplace=True)
```

```python
df_2020
```

```python
df_2020['PUNTO DE VENTA'].value_counts()
```

```python
#Groupby sales per month and channel, adding the values of the total with sum
df_2020 = df_2020.groupby(['MONTH','PUNTO DE VENTA'])['TOTAL'].sum().reset_index()
```

```python
#Adding '-2020' to enable plot
df_2020 = df_2020.set_index('MONTH')
df_2020 = df_2020.T.add_suffix('-2020').T
df_2020.reset_index()
```

```python
#Splitting dataframes per channel
df1_2020 = df_2020[df_2020['PUNTO DE VENTA'] == 'SHOWROOM']
df2_2020 = df_2020[df_2020['PUNTO DE VENTA'] == 'SHOPIFY']
```

```python
df1_2020
```

```python
df2_2020
```

```python
#Replacing names with mapping for consistency accross datasets
df1_2020.drop(['PUNTO DE VENTA'],axis=1,inplace=True)
df1_2020 = df1_2020.rename(columns={'TOTAL':'Ventas Menudeo', 'MONTH':'Month'})

df2_2020.drop(['PUNTO DE VENTA'],axis=1,inplace=True)
df2_2020 = df2_2020.rename(columns={'TOTAL':'Ventas en línea','MONTH':'Month' })
```

```python
df1_2020.reset_index()
```

```python
df2_2020.reset_index()
```

```python
#Merging dataframes by month
df_20_merged = pd.merge(df1_2020, df2_2020,
                        how="outer", on=["MONTH"])
df_20_merged
```

```python
#Filling NaN value
df_20_merged = df_20_merged.fillna(0)
df_20_merged
```

```python
df_20_merged = df_20_merged.reset_index()
df_20_merged
```

```python
df_20_merged= df_20_merged.reindex([1,2,4,0,5,3])
df_20_merged
```

```python
#Renaming columns for consistency
df_total_ = df_total_[['Month','Ventas Menudeo', 'Ventas en línea']]
df_total_ = df_total_.rename(columns={'Month':'MONTH'})
df_total_
```

```python
#concatenating files for 2018 to 2020
df_sales=pd.concat([df_total_, df_20_merged])
#df_sales = df_sales.reset_index(drop=True)
df_sales
```

```python
#Plotting SALES (seaborn)
#To activate or deactivate different sale channels remove or type '#' in front of each code line

from matplotlib.lines import Line2D

fig = plt.figure(figsize=(16,4))

ax = fig.add_subplot(111)
sns.set_theme(style="whitegrid")
sns.set_style("whitegrid", {'axes.grid' : False})

#Defining plot colors
colors_pres = ['#483D8B','#4169E1','#87CEFA','#808080' ,'#5F9EA0','#6495ED','#ADD8E6','#7B68EE']

#Defining lines
ax = sns.lineplot(x='MONTH',y=df_sales['Ventas Menudeo'],data=df_sales,lw=4,color=colors_pres[0],ax=ax)

ax = sns.lineplot(x='MONTH',y=df_sales['Ventas en línea'],data=df_sales,lw=4,color=colors_pres[4],ax=ax)


plt.title('Sales Analysis                                           ',fontsize=24,ha='right',va='top',color='#808080', y=1)
#sns.despine()

# Change the colors of the left and bottom borders (fade into the background)
ax.spines['left'].set_color('lightgray')
ax.spines['bottom'].set_color('lightgray')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


plt.ylabel('MXN Pesos',fontsize=12,color='#616161',ha='left',va='bottom')


legend_elements= [Line2D([0], [0], color=colors_pres[0], label='Ventas Menudeo',
                         markersize=10),
                   #Line2D([0], [0], color=colors_pres[1], label='Ventas Mayoreo',
                    #       markersize=10),
                #Line2D([0], [0], color=colors_pres[2], label='Ventas Ferias de Diseño',
                 #         markersize=10),
                 #Line2D([0], [0], color=colors_pres[3], label='Ventas Consignación',
                  #        markersize=10),
                  Line2D([0], [0], color=colors_pres[4], label='Ventas En Linea',
                          markersize=10)
                 ]


ax.legend(handles=legend_elements)
#ax.get_legend().remove()
#ax.legend(handles=legend_elements)
plt.xticks(rotation='vertical')
plt.show()
fig.savefig('Sales_Analysis_1.png',bbox_inches="tight")
```
