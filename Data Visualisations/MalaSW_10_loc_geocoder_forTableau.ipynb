```python
from geopy.geocoders import Nominatim
import numpy as np
import pandas as pd
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
#Locating cities where Mala is present using Geocoder for plotting on map using Tableau
```

```python
def city_locate(df):
    '''
    Inputs city. Returns the lat/long coordinates of 
    either the city if possible, if not, then returns lat/long of the center of the country.
    '''
    df['Latitude'] = None
    df['Longitude'] = None
    locator = Nominatim(user_agent="myGeocoder")
    
    for i in range(len(df)):
        location = locator.geocode(df.loc[i,'name'])
        df.loc[i,'Latitude'] = location.latitude   
        df.loc[i,'Longitude'] = location.longitude
        
    return df 
```

```python
# Locating cities with stores with wholesale agreements
cities = ['Casitas', 'Maraika', 'Punta de Mita', 'Holbox', 'Oaxaca',
          'San Miguel de Allende', 'Cabo', 'Nayarit', 'León']
cities = pd.DataFrame(cities)
cities.columns = ['name']
```

```python
new_cities = city_locate(cities)
```

```python
#Adding type column
new_cities['Type']= 'Mayoreo'
```

```python
new_cities
```

```python
# Locating cities with stores with consignment agreements
cities2 = ['Guadalajara', 'Providencia', 'Giovanna', 'Ciudad de México','Ciudad de México', 'Ciudad de México',
           'Ciudad Mante', 'Puebla', 'León', 'Ciudad de México', 'Ciudad de México']
cities2 = pd.DataFrame(cities2)
cities2.columns = ['name']
```

```python
new_cities_consig = city_locate(cities2)
```

```python
#Adding type column
new_cities_consig['Type']= 'Consigna'
```

```python
new_cities_consig
```

```python
#Concatenating both types
df_locations = pd.concat([new_cities, new_cities_consig], ignore_index=True)
```
